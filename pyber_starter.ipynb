```python
%matplotlib inline
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# Read the City and Ride Data

city_data = pd.read_csv("data/city_data.csv")
ride_data= pd.read_csv("data/ride_data.csv")
```

```python
# Combine the data into a single dataset

combine_df = pd.merge(ride_data, city_data, on="city", how= "outer")

# Display the data table for preview

combine_df.head()
```

```python
# Rename the Columns


combine_df = combine_df.rename(columns={"city": "City", "date":"Date", "fare": "Fare", "ride_id": "Ride ID", "driver_count": "Driver Count","type": "Type"})

combine_df.head(45)
```

```python
# Obtain the city types

cities= combine_df["Type"].unique()
cities
```

```python
# Obtain the total number of rides per City Type

grouped_city_type_df= combine_df.groupby(["Type"])

grouped_city_type_df.count()["City"]
```

```python
# Obtain the average fare per city and per ride

average_fare_city_type_df= combine_df.groupby(["Type"]).mean()

average_fare_per_ride_df= combine_df.groupby(["Type"]).mean()
```

```python
# Build the scatter plots for each city types
# ver tamano de las burbujas!

urban_loc = combine_df.loc[combine_df["Type"] == "Urban", :]

urban_x = urban_loc.groupby("City").count()["Ride ID"] 
urban_y = urban_loc.groupby("City").mean()["Fare"]
urban_z = urban_loc["Driver Count"]*10



rural_loc = combine_df.loc[combine_df["Type"] == "Rural", :]

rural_x = rural_loc.groupby("City").count()["Ride ID"] 
rural_y = rural_loc.groupby("City").mean()["Fare"]
rural_z = rural_loc["Driver Count"]*10



suburban_loc = combine_df.loc[combine_df["Type"] == "Suburban", :]

suburban_x = suburban_loc.groupby("City").count()["Ride ID"] 
suburban_y = suburban_loc.groupby("City").mean()["Fare"]
suburban_z = suburban_loc["Driver Count"]*10

```

```python
# Incorporate the other graph properties

plt.scatter(urban_x, urban_y, alpha=0.8, label="Urban", facecolors="coral", edgecolors="black", s= urban_z)
plt.scatter(rural_x, rural_y, alpha=0.8, label="Rural", facecolors="gold", edgecolors="black", s= rural_z)
plt.scatter(suburban_x, suburban_y, alpha=0.8 , label="Suburban", facecolors="lightskyblue", edgecolors="black", s= suburban_z)

plt.xlabel("Total Number of Rides (Per City)")
plt.ylabel("Average Fare ($)")
plt.title("Pyber Ride Sharing Data (2016)")
plt.grid()
plt.legend(fontsize = 10, title= "City Types", loc="best" )

# To set size of legend points
from matplotlib.collections import PathCollection
from matplotlib.legend_handler import HandlerPathCollection

def update(handle, orig):
    handle.update_from(orig)
    handle.set_sizes([64])
plt.legend(loc="upper right", prop={'size': 10},handler_map={PathCollection : HandlerPathCollection(update_func=update)} )


# Save the figure

plt.savefig("../Images/Pyber Ride Sharing Data.png")
```

## Total Fares by City Type

```python
# Calculate the percents of Rural - Urban and Suburban

fare_percent =  (combine_df.groupby("Type").sum()["Fare"]   / combine_df.sum()["Fare"])* 100 

# fare_percent (print to verify the values)


# Build Pie Chart

labels = ["Rural","Suburban","Urban"]
colors = ['gold','lightskyblue','lightcoral']
explode = (0, 0, 0.07)
plt.pie(fare_percent, explode=explode, labels=labels,autopct="%1.1f%%",colors=colors, shadow=True, startangle=140)
plt.axis("tight")
plt.title("% Total Fares by City Type")

# Save Figure
plt.savefig("../Images/Total Fares by City Type.png")

# Show Figure
plt.show()

```

## Total Rides by City Type

```python
# Calculate Ride Percents

rides_percent =  (combine_df.groupby("Type").count()["Date"]   / combine_df.count()["Date"])* 100 

# rides_percent (print to verify the values)


# Build Pie Chart

labels = ["Rural","Suburban","Urban"]
colors = ['gold','lightskyblue','lightcoral']
explode = (0, 0, 0.07)
plt.pie(rides_percent, explode=explode, labels=labels,autopct="%1.1f%%",colors=colors, shadow=True, startangle=140)
plt.axis("tight")
plt.title("% Total Rides by City Type")

# Save Figure
plt.savefig("../Images/Total Rides by City Type.png")


# Show Figure
plt.show()
```
